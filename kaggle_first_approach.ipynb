```
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/My Drive/Проект 'рекомендательная система для магазина'/data.csv")
```

```
Invoices = df['Invoice'].unique()
indexes = []
for i in range(len(Invoices)):
    if Invoices[i][0] == 'C':
        indexes.append(i)
Invoices_n = np.delete(Invoices, indexes)
df = df.loc[df['Invoice'].isin(Invoices_n)]
df_c = df[['StockCode', 'Customer ID']].copy()
df_c = df_c.dropna()
df_c = df_c.reset_index(drop = True)
print(df_c)
```

```
dict = {}
for i in df_c['Customer ID'].unique():
    dict[i] = (df_c[df_c['Customer ID'] == i]['StockCode']).tolist()
print(len(df_c['StockCode'].unique()))
print(len(df_c['Customer ID'].unique()))
print(len(dict))
dict
```

```
vects = np.zeros((5881, 4631))
```

```
name = df_c['Customer ID'].unique()
product = df_c.StockCode.unique()
vect_matrix = pd.DataFrame(data=vects, index=name, columns=product)
vect_matrix
```

```
import time
import datetime
start_time = datetime.datetime.now()
length = len(df_c)
#Итерируем по всем покупкам
#Добавляем единицу купленного товара пользователю 'Customer ID' по столбцу 'StockCode'
for i in range(length):
    vect_matrix.loc[df_c.loc[i, 'Customer ID'], df_c.loc[i, 'StockCode']] += 1
print('Time elapsed:', datetime.datetime.now() - start_time)
vect_matrix
```

```
one_buyers=[]
last_thing=[]
for i in dict.keys():
    if len(dict[i])==1:
        one_buyers.append(i)
for i in one_buyers:
    del dict[i]
for i in dict.keys():
    last_thing.append(dict[i][-1])
print(len(one_buyers))
print(len(last_thing))
```

```
vect_matrix=vect_matrix.drop(one_buyers)
vect_matrix
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(vect_matrix, last_thing, test_size=0.33, random_state=42)
```

```
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn import metrics
import time
import datetime
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

start_time = datetime.datetime.now()
fold = KFold(n_splits=5, shuffle=True, random_state=241)
parameters = {'estimator__max_depth': [10]}
clf = OneVsRestClassifier(DecisionTreeClassifier(random_state=241))
print(clf.get_params().keys())
print(sorted(metrics.SCORERS.keys()))
grid = GridSearchCV(clf, param_grid=parameters, scoring='accuracy', cv=fold)
grid.fit(X_train, y_train)
print('gs.best_score_:', grid.best_score_, '\n', 'gs.best_parameters_:', grid.best_params_)
print('Time elapsed:', datetime.datetime.now() - start_time)
print("Accuracy на тесте:",grid.score(X_test, y_test))
```

```
proba = grid.predict_proba(X_test)
start_time = datetime.datetime.now()
pos = 0
j=0
for i in proba:
  index = np.argpartition(i, -5)[-5:]
  classes = grid.classes_[index]
  if y_test[j] in classes:
    pos += 1
  j += 1
accuracy = pos/len(y_test)
print('Time elapsed:', datetime.datetime.now() - start_time)
print("accuracy:",accuracy)
```
